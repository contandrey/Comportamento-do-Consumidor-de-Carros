###**2.3 Padronização de dados**

Os dados geralmente são coletados de diferentes agências em diferentes formatos. (A padronização de dados também é um termo para um tipo específico de normalização de dados em que subtraímos a média e dividimos pelo desvio padrão.)

O que é padronização?

A padronização é o processo de transformar os dados em um formato comum, permitindo ao pesquisador fazer uma comparação significativa.

Padronização: A padronização é uma técnica de pré-processamento que transforma os dados para que tenham média igual a 0 e desvio padrão igual a 1. Isso facilita a comparação entre diferentes colunas e a convergência de modelos de aprendizado de máquina.

Desvio padrão após padronização: O desvio padrão de uma coluna padronizada é sempre 1. Isso ocorre porque a padronização subtrai a média e divide por desvio padrão original, resultando em uma nova coluna com média 0 e desvio padrão 1.

###**2.4 Normalização de dados**

Por que normalização?

Normalização é o processo de transformar valores de diversas variáveis ​​em um intervalo semelhante. Normalizações típicas incluem dimensionar a variável para que a média da variável seja 0, dimensionar a variável para que a variância seja 1 ou dimensionar a variável para que os valores da variável variem de 0 a 1.

Exemplo

Para demonstrar a normalização, digamos que queremos dimensionar as colunas “comprimento”, “largura” e “altura”.

Alvo: gostaria de normalizar essas variáveis ​​para que seus valores variassem de 0 a 1

Abordagem: substitua o valor original por (valor original)/(valor máximo)

```python
# substitua (valor original) por (valor original)/(valor máximo)
df['comprimento'] = df['comprimento']/df['comprimento'].max()
df['largura'] = df['largura']/df['largura'].max()
```


Normalizando a coluna “altura”.

```python
df['altura'] = df['altura']/df['altura'].max()

# show the scaled columns
df[["comprimento","largura","altura"]].head()
```

###**3.4 Correlação e Causalidade**
Correlação: uma medida da extensão da interdependência entre variáveis.

Causalidade: a relação entre causa e efeito entre duas variáveis.

É importante saber a diferença entre esses dois. Correlação não implica em causa. Determinar a correlação é muito mais simples do que determinar a causalidade, pois a causalidade pode exigir experimentação independente.

###**4.5.1 Correlação de Pearson**

A Correlação de Pearson mede a dependência linear entre duas variáveis ​​X e Y.

O coeficiente resultante é um valor entre -1 e 1 inclusive, onde:

1: Correlação linear positiva perfeita.
0: Sem correlação linear, as duas variáveis ​​provavelmente não afetam uma à outra.
-1: Correlação linear negativa perfeita.

Correlação de Pearson é o método padrão da função "corr". Como antes, podemos calcular a Correlação de Pearson das variáveis ​​'int64' ou 'float64'.

```python
df.corr()
```

Às vezes gostaríamos de saber a significância da estimativa de correlação.

Valor P

Qual é esse valor P? O valor P é o valor de probabilidade de que a correlação entre essas duas variáveis ​​seja estatisticamente significativa. Normalmente escolhemos um nível de significância de 0,05, o que significa que temos 95% de confiança de que a correlação entre as variáveis ​​é significativa.

Por convenção, quando o

valor p é < 0,001: dizemos que há fortes evidências de que a correlação é significativa.

o valor p é < 0,05: há evidência moderada de que a correlação é significativa.

o valor p é < 0,1: há fraca evidência de que a correlação é significativa.

o valor p é > 0,1: não há evidências de que a correlação seja significativa.

###**3.5 ANOVA**
ANOVA: Análise de Variância

A Análise de Variância (ANOVA) é um método estatístico utilizado para testar se existem diferenças significativas entre as médias de dois ou mais grupos. ANOVA retorna dois parâmetros:

Pontuação do teste F: ANOVA assume que as médias de todos os grupos são iguais, calcula o quanto as médias reais se desviam da suposição e relata isso como a pontuação do teste F. Uma pontuação maior significa que há uma diferença maior entre as médias.

Valor P: O valor P indica o quão estatisticamente significativo é o nosso valor de pontuação calculado.

Se nossa variável de preço estiver fortemente correlacionada com a variável que estamos analisando, esperamos que a ANOVA retorne uma pontuação considerável no teste F e um valor p pequeno.

Função linear

Gráfico Residual

Uma boa maneira de visualizar a variância dos dados é usar um gráfico residual.

O que é um resíduo?

A diferença entre o valor observado (y) e o valor previsto (Yhat) é chamada de resíduo (e). Quando olhamos para um gráfico de regressão, o resíduo é a distância do ponto de dados até a linha de regressão ajustada.

Então, o que é um gráfico residual?

Um gráfico residual é um gráfico que mostra os resíduos no eixo y vertical e a variável independente no eixo x horizontal.

Em que prestamos atenção quando olhamos para um gráfico residual?

Observamos a propagação dos resíduos:

- Se os pontos num gráfico residual estiverem distribuídos aleatoriamente em torno do eixo x, então um modelo linear é apropriado para os dados.

Por que é que? Resíduos distribuídos aleatoriamente significam que a variância é constante e, portanto, o modelo linear é um bom ajuste para esses dados.

```python
width = 12
height = 10
plt.figure(figsize=(width, height))
sns.residplot(x=df['rodovia-mpg'],y=df['preço'])
plt.show()
```

O que esse enredo está nos dizendo?

Podemos ver neste gráfico de resíduos que os resíduos não estão espalhados aleatoriamente em torno do eixo x, levando-nos a acreditar que talvez um modelo não linear seja mais apropriado para estes dados.

###**4.2 Regressão Polinomial e Pipelines**

A regressão polinomial é um caso particular do modelo de regressão linear geral ou modelos de regressão linear múltipla.

Obtemos relacionamentos não lineares elevando ao quadrado ou definindo termos de ordem superior das variáveis ​​preditoras.

Cúbico - 3ª Ordem

Ordem superior:

Vimos anteriormente que um modelo linear não forneceu o melhor ajuste ao usar "highway-mpg" como variável preditora. Vamos ver se podemos tentar ajustar um modelo polinomial aos dados.

Usaremos a seguinte função para plotar os dados:

```python
def PlotPolly(model, independent_variable, dependent_variabble, Name):
    x_new = np.linspace(15, 55, 100)
    y_new = model(x_new)

    plt.plot(independent_variable, dependent_variabble, '.', x_new, y_new, '-')
    plt.title('Ajuste polinomial com Matplotlib para preço ~ comprimento')
    ax = plt.gca()
    ax.set_facecolor((0.898, 0.898, 0.898))
    fig = plt.gcf()
    plt.xlabel(Name)
    plt.ylabel('Preço de carros')

    plt.show()
    plt.close()
```

```python
# Aqui usamos um polinômio de 3ª ordem (cúbico)
f = np.polyfit(x, y, 3)
p = np.poly1d(f)
print(p)
```

Vamos traçar a função:

```python
PlotPolly(p, x, y, 'rodovia-mpg')
```

R ao quadrado, também conhecido como coeficiente de determinação, é uma medida que indica o quão próximos os dados estão da linha de regressão ajustada.

Erro Quadrático Médio (MSE)

```python
def DistributionPlot(RedFunction, BlueFunction, RedName, BlueName, Title):
    width = 12
    height = 10
    plt.figure(figsize=(width, height))

    ax1 = sns.kdeplot(RedFunction, color="r", label=RedName)
    ax2 = sns.kdeplot(BlueFunction, color="b", label=BlueName, ax=ax1)

    plt.title(Title)
    plt.xlabel('Preço em dólar)')
    plt.ylabel('Proporção de carros')
    plt.show()
    plt.close()
```

```python
def PollyPlot(xtrain, xtest, y_train, y_test, lr,poly_transform):
    width = 12
    height = 10
    plt.figure(figsize=(width, height))


    #Dados de treinamento
    #Dados de teste
    # lr:  objeto de regressão linear
    #poly_transform:  objeto de transformação polynomial

    xmax=max([xtrain.values.max(), xtest.values.max()])

    xmin=min([xtrain.values.min(), xtest.values.min()])

    x=np.arange(xmin, xmax, 0.1)


    plt.plot(xtrain, y_train, 'ro', label='Dados de treinamento')
    plt.plot(xtest, y_test, 'go', label='Dados de teste')
    plt.plot(x, lr.predict(poly_transform.fit_transform(x.reshape(-1, 1))), label='Função prevista')
    plt.ylim([-10000, 60000])
    plt.ylabel('Preço')
    plt.legend()
```

```python
from tqdm import tqdm

Rsqu_test = []
Rsqu_train = []
dummy1 = []
Alpha = 10 * np.array(range(0,1000))
pbar = tqdm(Alpha)

for alpha in pbar:
    RigeModel = Ridge(alpha=alpha)
    RigeModel.fit(x_train_pr, y_train)
    test_score, train_score = RigeModel.score(x_test_pr, y_test), RigeModel.score(x_train_pr, y_train)

    pbar.set_postfix({"Pontuação do teste": test_score, "Pontuação do train": train_score})

    Rsqu_test.append(test_score)
    Rsqu_train.append(train_score)
```
